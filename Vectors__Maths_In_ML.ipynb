```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,6,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-2,6)
plt.ylim(-2,6)
plt.show()
```

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,-1,2])
```

```python
sum1 = vector1 + vector2
print(sum1)
```

```python
plt.quiver(','.join(vector1),scale_units='xy',angles='xy',scale = 1, color='y')
plt.quiver(vector2,scale_units='xy',angles='xy',scale = 1, color='g')
plt.quiver(sum1,scale_units='xy',angles='xy',scale=1,color='r')
plt.show()
```

# Dot Product of 2 Vectors 

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b) #calculating dot product
```

```python
a_dot_b
```

```python
c = np.array([40,20,35])
d =  np.array([53,24,68])
```

```python
c_dot_d = np.dot(c,d)
```

```python
c_dot_d
```

# Cross Product

```python
a = np.array([2,3])
b = np.array([4,4])

a_cross_b = np.cross(a,b)
a_cross_b
```

```python
a = np.array([5,3,10])
b = np.array([10,-4,20])

a_cross_b = np.cross(a,b)
a_cross_b
```
